```python
import pandas as pd
import time
import os

start_time = time.time()

genre = 'k-pop'
timestamp = time.strftime("%Y%m%d-%H%M%S")

artist_id = []
artist_name = []
artist_popularity = []
artist_followers = []

track_id = []
track_name = []
release_date = []
track_popularity = []

album_image = []

track_features_list = []

iteration = 0  

for i in range(0, 1000, 50):
    iteration += 1  
    track_results = sp.search(q=f'year:2023 genre:"{genre}"', type='track', limit=5, offset=i)
    for j, t in enumerate(track_results['tracks']['items']):
        artist = sp.artist(t['artists'][0]['id'])

        artist_id.append(artist['id'])
        artist_name.append(artist['name'])
        artist_popularity.append(artist['popularity'])
        artist_followers.append(artist['followers']['total'])

        track_id.append(t['id'])
        track_name.append(t['name'])
        release_date.append(t['album']['release_date'])
        track_popularity.append(t['popularity'])

        try:
            album_image.append(t['album']['images'][1]['url']) 
        except IndexError:
            album_image.append(None)  

        print(f"Processing track {j+1} by {artist['name']}...") 

        track_features = sp.audio_features([t['id']])
        if track_features:
            track_features_list.append(track_features[0])
        else:
            track_features_list.append(None)

        #time.sleep(360) 

   
    artist_df = pd.DataFrame(
        {'artist_id': artist_id, 'artist_name': artist_name, 'artist_popularity': artist_popularity,
         'artist_followers': artist_followers})
    artist_df.to_csv(f'/home/ubuntu/project/4.5HZ/Data/API_TOP/TOP_artist/{genre}_artist_{timestamp}_{iteration}.csv', index=False)  # ÏàòÏ†ïÎêú Î∂ÄÎ∂Ñ: ÌååÏùº Ïù¥Î¶ÑÏóê iteration Ï∂îÍ∞Ä

  
    track_df = pd.DataFrame(
        {'track_id': track_id, 'track_name': track_name, 'release_date': release_date, 'track_popularity': track_popularity,
         'album_image': album_image})
    track_df.to_csv(f'/home/ubuntu/project/4.5HZ/Data/API_TOP/TOP_track/{genre}_track_{timestamp}_{iteration}.csv', index=False)  # ÏàòÏ†ïÎêú Î∂ÄÎ∂Ñ: ÌååÏùº Ïù¥Î¶ÑÏóê iteration Ï∂îÍ∞Ä

    track_features_df = pd.DataFrame(track_features_list)
    track_features_df = track_features_df.drop(columns=['type', 'id', 'uri', 'track_href', 'analysis_url', 'duration_ms', 'time_signature'])

    track_features_df.to_csv(f'/home/ubuntu/project/4.5HZ/Data/API_TOP/TOP_audio/{genre}_audio_features_{timestamp}_{iteration}.csv', index=False)  # ÏàòÏ†ïÎêú Î∂ÄÎ∂Ñ: ÌååÏùº Ïù¥Î¶ÑÏóê iteration Ï∂îÍ∞Ä

    time.sleep(60)  

end_time = time.time()
print(f"ÏûëÏóÖ ÏÜåÏöî ÏãúÍ∞Ñ: {end_time - start_time} Ï¥à")
```
