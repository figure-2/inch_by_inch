```python
# 수정전 코드 (백업)
start_time = time.time()

track_id = []
track_name = []
release_date = []
track_popularity = []
album_image = []

artist_id = []
artist_name = []
artist_genres = []

for i in range(0, 5, 50):
    track_results = sp.search(q=f'year:2023 genre:"{genre}"', type='track', limit=50, offset=i)
    for j, t in enumerate(track_results['tracks']['items']):
        artist = sp.artist(t['artists'][0]['id'])
        
        artist_id.append(artist['id'])
        artist_name.append(artist['name'])
        artist_genres.append(artist['genres'])

        track_id.append(t['id'])
        track_name.append(t['name'])
        release_date.append(t['album']['release_date'])
        track_popularity.append(t['popularity'])

        try:
            album_image.append(t['album']['images'][1]['url'])  # 320 사이즈 이미지
        except IndexError:
            album_image.append(None)  

        print(f"Processing track {j+1}...")  

    #time.sleep(60)  
    
track_df = pd.DataFrame(
    {'artist_id': artist_id, 'artist_name': artist_name,'track_id': track_id, 'track_name': track_name, 'release_date': release_date, 'artist_genres': artist_genres, 'track_popularity': track_popularity,
     'album_image': album_image})
track_df.to_csv(f'/home/ubuntu/project/4.5HZ/Data/API_TOP/TOP_track/{genre}_track_{timestamp}.csv', index=False)

end_time = time.time()
print(f"작업 소요 시간: {end_time - start_time} 초")
```

```python
# 수정전 코드 (백업)
start_time = time.time()

print("오디오 피쳐 6분 대기")
time.sleep(10)

track_features_list = []

for i in range(0, 3, 50):
    track_results = sp.search(q=f'year:2023 genre:"{genre}"', type='track', limit=50, offset=i)
    for j, t in enumerate(track_results['tracks']['items']):
        track_features = sp.audio_features([t['id']])
        if track_features:
            track_features_list.append(track_features[0])
        else:
            track_features_list.append(None)

        print(f"Processing audio features for track {j+1}...")  

    #time.sleep(360)  

track_features_df = pd.DataFrame(track_features_list)
track_features_df = track_features_df.rename(columns={'id': 'track_id'})
track_features_df = track_features_df.drop(columns=['type', 'uri', 'track_href', 'analysis_url', 'duration_ms', 'time_signature'])

track_features_df.to_csv(f'/home/ubuntu/project/4.5HZ/Data/API_TOP/TOP_audio/{genre}_audio_22features_{timestamp}.csv', index=False)

end_time = time.time()
print(f"작업 소요 시간: {end_time - start_time} 초")
```

```python
import pandas as pd
import time
from spotipy import SpotifyException

start_time = time.time()

genre_list = ['k-pop', 'acoustic', 'club', 'dance', 'edm', 'disco', 'electronic', 'happy', 'hip-hop', 'jazz', 'party', 'pop', 'r-n-b', 'sad', 'rock', 'romance', 'sleep', 'study', 'soul', 'kids', 'classical']
timestamp = time.strftime("%Y%m%d-%H%M%S")

track_features_list = []

for genre in genre_list:
    print(f"{genre}: 오디오 피쳐 다운로드 전 6분 대기.")
    time.sleep(5)

    for i in range(0, 3, 50):
        track_results = sp.search(q=f'year:2023 genre:"{genre}"', type='track', limit=50, offset=i)
        for j, t in enumerate(track_results['tracks']['items']):
            time.sleep(10)  # 각 항목에 대한 딜레이
            try:
                track_features = sp.audio_features([t['id']])
                if track_features:
                    track_features_list.append(track_features[0])
                else:
                    track_features_list.append(None)
            except SpotifyException as e:
                if e.http_status == 429:
                    print("API 요청 제한 초과. 대기 중...")
                    time.sleep(5)  # API 제한 초과 시 더 긴 대기 시간

            print(f"Processing audio features for track {j+1}...")  

# 나머지 코드는 동일

track_features_df = pd.DataFrame(track_features_list)
track_features_df = track_features_df.rename(columns={'id': 'track_id'})
track_features_df = track_features_df.drop(columns=['type', 'uri', 'track_href', 'analysis_url', 'duration_ms', 'time_signature'])

track_features_df.to_csv(f'/home/ubuntu/project/4.5HZ/Data/API_TOP/TOP_audio/audio_22features_{timestamp}.csv', index=False)

end_time = time.time()
print(f"작업 소요 시간: {end_time - start_time} 초")
```

플레이리스트 가져오는 코드

```python
import pandas as pd
import time
import os
import datetime

start_time = time.time()

search_query = 'k-pop'
search_results = sp.search(search_query, type='playlist', limit=50)

playlist_ids = [playlist['id'] for playlist in search_results['playlists']['items']]

data = []
for playlist_id in playlist_ids:
    results = sp.playlist(playlist_id)
    for idx, item in enumerate(results['tracks']['items']):
        track = item['track']
        if track is None:
            continue
        data.append([results['id'], results['name'], results['owner']['id'], track['artists'][0]['id'], track['artists'][0]['name'], track['id'], track['name']])

df = pd.DataFrame(data, columns=['playlist_id', 'playlist_name', 'user_id', 'Artist ID', 'Artist', 'Track ID', 'Track'])

timestamp = datetime.datetime.now().strftime("%Y%m%d_%H%M%S") 
df.to_csv(f'/home/ubuntu/project/4.5HZ/Data/API_playlist/playlist_info/{search_query}_audio_features_{timestamp}.csv', index=False, encoding='utf-8-sig')

end_time = time.time() 
print(f"작업 시작 시간: {start_time}") 
print(f"작업 종료 시간: {end_time}")  
print(f"작업 소요 시간: {end_time - start_time} 초") 
```
