```python
import pandas as pd


# Set up your query parameters
genre = 'k-pop'
limit = 5

# Get the top tracks in the K-pop genre
top_tracks = sp.search(q=f'genre:"{genre}"', type='track', limit=limit)

# Initialize empty lists for track features and audio analysis
track_features_list = []
audio_analysis_list = []

# Iterate over the top tracks and retrieve track features and audio analysis
for track in top_tracks['tracks']['items']:
    track_id = track['id']
    
    # Get track features
    track_features = sp.audio_features(track_id)[0]
    track_features_list.append(track_features)
    
    # Get audio analysis
    audio_analysis = sp.audio_analysis(track_id)
    
    # Flatten the audio analysis dictionary and append to the list
    flattened_audio_analysis = {}
    for key, value in audio_analysis.items():
        if isinstance(value, dict):
            for subkey, subvalue in value.items():
                flattened_audio_analysis[f"{key}_{subkey}"] = subvalue
        else:
            flattened_audio_analysis[key] = value
    
    audio_analysis_list.append(flattened_audio_analysis)

# Create DataFrames for track features and flattened audio analysis
track_features_df = pd.DataFrame(track_features_list)
audio_analysis_df = pd.DataFrame(audio_analysis_list)

# Combine track features and audio analysis DataFrames
combined_df = pd.concat([track_features_df, audio_analysis_df], axis=1)

# Save DataFrame to CSV file
combined_df.to_csv('./data/k-pop_10002222.csv', index=False)
```

```python
import pandas as pd
import time
start_time = time.time()

artist_id =[]
artist_name =[]
artist_popularity = []
artist_followers = []

track_id =[]
track_name = []
release_date = []
artist_genres = []
track_popularity =[]

album_image = []

track_features_list = []


for i in range(0,5,50):
    track_results = sp.search(q='year:2023 genre:"k-pop"', type='track', limit=5, offset=i)
    for i, t in enumerate(track_results['tracks']['items']):
        artist = sp.artist(t['artists'][0]['id'])

        artist_id.append(artist['id'])
        artist_name.append(artist['name'])
        artist_genres.append(artist['genres'])
        artist_popularity.append(artist['popularity'])
        artist_followers.append(artist['followers']['total'])

        track_id.append(t['id'])
        track_name.append(t['name'])
        release_date.append(t['album']['release_date'])
        track_popularity.append(t['popularity'])

        try:
            album_image.append(t['album']['images'][1]['url'])  # 320x320 사이즈 이미지
        except IndexError:
            album_image.append(None)  # If there is no image or only one image
    
        track_features = sp.audio_features([t['id']])
        if track_features:
            track_features_list.append(track_features[0])
        else:
            track_features_list.append(None)

Spotify_Top_tracks = pd.DataFrame(
    {'artist_id' : artist_id, 'artist_name' : artist_name, 'artist_popularity' : artist_popularity, 'artist_followers' : artist_followers,
     'track_id' : track_id, 'track_name' : track_name, 'release_date' : release_date, 'artist_genres' : artist_genres, 'track_popularity' : track_popularity,
     'album_image' : album_image})

track_features_df = pd.DataFrame(track_features_list)


final_df = pd.concat([Spotify_Top_tracks, track_features_df], axis=1)

end_time = time.time()
print(f"Spotify_Top_tracks 작업 소요 시간: {end_time - start_time} 초")

final_df.head()


final_df.to_csv('./Spotify_Top_tracks.csv', index=False)
```
