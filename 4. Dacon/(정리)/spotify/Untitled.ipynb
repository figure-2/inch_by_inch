```python
import pandas as pd
import time
import os

start_time = time.time()

genre = 'k-pop'
timestamp = time.strftime("%Y%m%d-%H%M%S")

artist_id = []
artist_name = []
artist_popularity = []
artist_followers = []

track_id = []
track_name = []
release_date = []
track_popularity = []

album_image = []

track_features_list = []



for i in range(0, 5, 50):
    track_results = sp.search(q=f'year:2023 genre:"{genre}"', type='track', limit=5, offset=i)
    for j, t in enumerate(track_results['tracks']['items']):
        artist = sp.artist(t['artists'][0]['id'])

        artist_id.append(artist['id'])
        artist_name.append(artist['name'])
        artist_popularity.append(artist['popularity'])
        artist_followers.append(artist['followers']['total'])

        track_id.append(t['id'])
        track_name.append(t['name'])
        release_date.append(t['album']['release_date'])
        track_popularity.append(t['popularity'])

        try:
            album_image.append(t['album']['images'][1]['url']) 
        except IndexError:
            album_image.append(None)  
        print(f"Processing track {j+1} by {artist['name']}...") 

        track_features = sp.audio_features([t['id']])
        if track_features:
            track_features_list.append(track_features[0])
        else:
            track_features_list.append(None)

        

    # Save artist data
    artist_df = pd.DataFrame(
        {'artist_id': artist_id, 'artist_name': artist_name, 'artist_popularity': artist_popularity,
         'artist_followers': artist_followers})
    artist_df.to_csv(f'/home/ubuntu/project/4.5HZ/Data/API_artist/{genre}_artist_{timestamp}.csv', index=False) 

    # Save track data
    track_df = pd.DataFrame(
        {'track_id': track_id, 'track_name': track_name, 'release_date': release_date, 'track_popularity': track_popularity,
         'album_image': album_image})
    track_df.to_csv(f'/home/ubuntu/project/4.5HZ/Data/API_track/{genre}_track_{timestamp}.csv', index=False)  

    # Save audio features data
    track_features_df = pd.DataFrame(track_features_list)
    track_features_df = track_features_df.drop(columns=['type', 'id', 'uri', 'track_href', 'analysis_url', 'duration_ms', 'time_signature'])

    track_features_df.to_csv(f'/home/ubuntu/project/4.5HZ/Data/API_audio/{genre}_audio_features_{timestamp}.csv', index=False)  

    time.sleep(60)  

end_time = time.time()
print(f"작업 소요 시간: {end_time - start_time} 초")

위에 코드 내용을 수정하고 싶어 csv로 저장 되는 파일 형식이 총 3개 있는데 
2개의 셀에서 따로따로 실행하고 싶어 
첫번쨰 셀에서는   아래 부분이 실행이 되고 
# Save artist data
    artist_df = pd.DataFrame(
        {'artist_id': artist_id, 'artist_name': artist_name, 'artist_popularity': artist_popularity,
         'artist_followers': artist_followers})
    artist_df.to_csv(f'/home/ubuntu/project/4.5HZ/Data/API_artist/{genre}_artist_{timestamp}.csv', index=False) 

    # Save track data
    track_df = pd.DataFrame(
        {'track_id': track_id, 'track_name': track_name, 'release_date': release_date, 'track_popularity': track_popularity,
         'album_image': album_image})
    track_df.to_csv(f'/home/ubuntu/project/4.5HZ/Data/API_track/{genre}_track_{timestamp}.csv', index=False)  
두번쨰 셀에서는 아래 부분이 실행되고 싶어

    # Save audio features data
    track_features_df = pd.DataFrame(track_features_list)
    track_features_df = track_features_df.drop(columns=['type', 'id', 'uri', 'track_href', 'analysis_url', 'duration_ms', 'time_signature'])

    track_features_df.to_csv(f'/home/ubuntu/project/4.5HZ/Data/API_audio/{genre}_audio_features_{timestamp}.csv', index=False)  
    

그리고 두 셀 모두 공통적으로 작업 시작시간 종료시간 작업 소요시간 그리고 진행상황을 보고싶어
```
