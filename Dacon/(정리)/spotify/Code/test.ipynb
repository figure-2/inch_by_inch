```python
import time
import pandas as pd

start_time = time.time()

year_list = ['2023', '2022', '2021', '2020']

for year in year_list:

    start_time = time.time()
    
    genre_list = ['k-pop', 'acoustic', 'dance', 'edm', 'disco', 
                'electronic', 'happy',  'hip-hop', 'jazz', 'party', 
                'pop', 'r-n-b', 'sad', 'rock', 'romance', 
                'sleep', 'study', 'soul', 'kids', 'classical']

    timestamp = time.strftime("%Y%m%d-%H%M%S")

    for genre in genre_list:
        
        print(f"{year}_{genre}: 다운로드 {end_time - start_time} 초")
        time.sleep(300)  
        
        track_features_list = []
        
        for i in range(0, 1000, 50):
            
            # print("1차 요청 대기")
            # time.sleep(60)
            
            track_results = sp.search(q=f'year:2023 genre:"{genre}"', type='track', limit=50, offset=i)
            for j, t in enumerate(track_results['tracks']['items']):
                
                # print("2차 요청 대기")
                # time.sleep(60)  
                
                track_features = sp.audio_features([t['id']])
                if track_features:
                    track_features_list.append(track_features[0])
                else:
                    track_features_list.append(None)
                print(f"Processing audio features for track {j+1}...")
                
                print("3차 요청 대기")
                time.sleep(60)  
            
        track_features_df = pd.DataFrame(track_features_list)
        track_features_df = track_features_df.rename(columns={'id': 'track_id'})
        track_features_df = track_features_df.drop(columns=['type', 'uri', 'track_href', 'analysis_url', 'duration_ms', 'time_signature'])
        track_features_df.to_csv(f'/home/ubuntu/project/4.5HZ/Data/API_TOP/TOP_audio/2020년대/{year}_{genre}_audio_features_{timestamp}.csv', index=False)
        
    end_time = time.time()
    print(f"작업 소요 시간: {end_time - start_time} 초")
```

```python
import time
import pandas as pd
from tqdm import tqdm 


genre_list = ['k-pop', 'acoustic', 'dance', 'edm', 'disco', 
              'electronic', 'happy', 'hip-hop', 'jazz', 'party', 
              'pop', 'r-n-b', 'sad', 'rock', 'romance', 
              'sleep', 'study', 'soul', 'kids', 'classical']


year_list = ['2023', '2022', '2021', '2020']

start_time = time.time()

for year in year_list:
    timestamp = time.strftime("%Y%m%d-%H%M%S")
    track_features_list = []

    for genre in genre_list:
        print(f"{year}_{genre}: 다운로드")
        time.sleep(300) 
        
        for i in tqdm(range(0, 1000, 50)):  
            track_results = sp.search(q=f'year:{year} genre:"{genre}"', type='track', limit=50, offset=i)
            for j, t in enumerate(track_results['tracks']['items']):
                track_features = sp.audio_features([t['id']])
                
                if track_features:
                    track_features_df = pd.DataFrame(track_features).rename(columns={'id': 'track_id'})
                    track_features_list.append(track_features_df)

                time.sleep(60)  

    
    all_track_features_df = pd.concat(track_features_list, ignore_index=True)

  
    output_path = f'/home/ubuntu/project/4.5HZ/Data/API_TOP/TOP_audio/2020년대/{year}_all_genres_audio_features_{timestamp}.csv'
    all_track_features_df.to_csv(output_path, index=False)

end_time = time.time()
print(f"전체 작업 소요 시간: {end_time - start_time} 초")
```
