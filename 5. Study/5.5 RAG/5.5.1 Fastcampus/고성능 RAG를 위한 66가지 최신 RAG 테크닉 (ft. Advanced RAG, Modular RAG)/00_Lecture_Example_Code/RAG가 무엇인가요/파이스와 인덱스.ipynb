```python
import numpy as np 
import faiss  # faiss 라이브러리 호출
```

```python
dimension = 128    # 각 벡터의 차원                
n = 200    # 벡터 수                   
np.random.seed(1)             
db_vectors = np.random.random((n, dimension)).astype('float32') #(200 * 128) 벡터 행렬

# 연습 예제라서 데이터는 임의의 행렬 데이터를 생성
```

```python
nlist = 5  # 클러스터(셀이라고도 표현) 수
quantiser = faiss.IndexFlatL2(dimension)  #quantizer를 활용해서 클러스터 생성 /IndexFlatL2를 이용
index = faiss.IndexIVFFlat(quantiser, dimension, nlist,   faiss.METRIC_L2)  #Inverted File 만들기 / 인덱스는 IndexIVFFlat를 이용
# quantiser : 클러스틑 만드는 부분, dimension : 긱 벡터의 차원, nlist : 클러스터의 수,   faiss.METRIC_L2 : 지정??

#index = faiss.IndexPQ(D, m, nbits)
#index = faiss.IndexIVFPQ(vecs, D, nlist, m, nbits)
```

```python
print(index.is_trained)   # False
index.train(db_vectors)  # 데이터베이스 벡터에 대한 훈련
print(index.ntotal)   
index.add(db_vectors)   # 벡터를 추가하고 인덱스를 업데이트 / 데이터를 인덱스에 넣어주는것 즉 추가 데이터를 넣어주는 것 / 인덱스를 재조합해서 훈련시켜주는 것
print(index.is_trained)  # True
print(index.ntotal)   
```

```python
#nprobe = 2  # 가장 유사한 클러스터 2개 찾기
n_query = 10  # 10개의 쿼리 벡터
k = 3  # 가장 가까운 이웃 3개를 반환
np.random.seed(1)  # 일관된 결과를 얻기 위해 시드 설정 
query_vectors = np.random.random((n_query, dimension)).astype('float32') #쿼리 벡터 생성 / 질문도 벡터로 생성
distances, indices = index.search(query_vectors, k) #검색 수행 , 검색 할 때 index.search를 사용하며 우리가 만들어준것을 사용 위에 셀에서
# index = faiss.IndexIVFFlat(quantiser, dimension, nlist,   faiss.METRIC_L2) 이것을 사용한다

# FAISS에서는 기본적으로 유사도 검색을 할때 Similarity Search를 사용하는데, 
# 지금 여기서는 임의로 만든 index를 search로 사용하는것


# 인덱스를 직접 만드는것이 어려움 / 인덱스에서 어떤것을 사용할지 고민해야 되고 train에서 하드웨어 소모가 필요하다
# 특히 index를 만들어서 사용하는것이 속도 및 정확도가 높은것의 보장이 없다
```
